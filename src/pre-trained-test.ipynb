```python
#Importing libraries
from ultralytics import YOLO
import cv2
import os
from glob import glob
```

```python
#Converting frames into a video for testing
frame_dir = "../data/VisDrone2019/val/sequences/uav0000137_00458_v"
output_video = "val_sequence.mp4"
fps = 24  # Adjust based on dataset

frame_paths = sorted(glob(os.path.join(frame_dir, "*.jpg")))
sample_frame = cv2.imread(frame_paths[0])
h, w, _ = sample_frame.shape

fourcc = cv2.VideoWriter_fourcc(*"mp4v")
video_out = cv2.VideoWriter(output_video, fourcc, fps, (w, h))

for path in frame_paths:
    img = cv2.imread(path)
    video_out.write(img)

video_out.release()
print(f"Saved video to: {output_video}")
```

```python
model = YOLO("yolov8s.pt")

# Track on the new video file (plays in one window, saves one output video)
results = model.track(
    source="val_sequence.mp4",
    tracker="bytetrack.yaml",
    show=True,
    save=True,
    verbose=True
)
```

```python
#Track the video file using Deep SORT
results = model.track(
    source="val_sequence.mp4",
    tracker="botsort.yaml",  # or deepsort.yaml if config is available
    show=True,
    save=True
)
```
